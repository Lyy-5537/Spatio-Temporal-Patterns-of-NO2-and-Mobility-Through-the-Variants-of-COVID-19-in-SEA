```python
import pandas as pd
```

```python
SI = pd.read_csv('https://raw.githubusercontent.com/liyangyang515/Spatio-Temporal-Patterns-of-NO2-and-Mobility-Through-the-Variants-of-COVID-19-in-SEA/main/data/stringency_index.csv')
SI = SI.rename({'Entity': 'country', 'Day':'date'}, axis='columns') 
SI
```

```python
case_num = pd.read_csv('https://raw.githubusercontent.com/liyangyang515/Spatio-Temporal-Patterns-of-NO2-and-Mobility-Through-the-Variants-of-COVID-19-in-SEA/main/data/new_cases.csv')
case_num = case_num[['location', 'date', 'new_cases']].rename({'location': 'country'}, axis='columns') 
case_num          
```

```python
merge = pd.merge(case_num, SI, how = 'inner', on = ['country', 'date'])
merge
```

```python
merge['date'] =  pd.to_datetime(merge['date'], format='%Y-%m-%d')
```

```python
merge['month'] = merge['date'].dt.month
merge['year'] = merge['date'].dt.year
merge
```

```python
# group by month, agg to mean
merge = merge.groupby(['year','month', 'country']).mean().reset_index()
merge
```

```python
merge_by_month_location = pd.read_csv(r'D:\GitHub\NO2-in-South-East-Asia-_GE5219\data\merge_by_month.csv', index_col = 0)
merge_by_month_location
```

```python
merge_all = pd.merge(merge_by_month_location, merge, how = 'left', on = ['year', 'month','country'])
merge_all
```

```python
merge_all.to_csv(r'D:\GitHub\NO2-in-South-East-Asia-_GE5219\data\merge_by_month_all.csv')
```
